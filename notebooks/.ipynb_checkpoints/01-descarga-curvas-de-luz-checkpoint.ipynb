```python
from lightkurve import search_lightcurvefile
import matplotlib.pyplot as plt

# Buscamos una estrella con trnsitos conocidos (por ejemplo, TIC 307210830)
target = "TIC 307210830"
lc_file = search_lightcurvefile(target, mission='TESS').download()
lc = lc_file.PDCSAP_FLUX.normalize().remove_nans()

# Guardamos en disco
lc.to_fits(path=f"../data/{target.replace(' ', '_')}.fits", overwrite=True)

# Visualizamos
lc.plot(title=f"Curva de luz de {target}")
plt.show()
```
